```python
from CGRtools import smiles, ReactionContainer
from itertools import chain
```

```python
bad = ['[OH:8][C:6](=[O:7])[CH:5]=[CH2:4].[CH3:1][C:2](=[CH2:3])[CH:18]=[CH2:17]>>[CH3:1][C:2]1=[CH:3][CH2:4][CH:5]([CH2:17][CH2:18]1)[C:6]([OH:8])=[O:7]',
       '[CH2:1]=[CH:2][CH2:3][O:17][c:16]1[cH:4][cH:5][cH:6][cH:14][cH:15]1>>[OH:17][c:16]1[cH:15][cH:14][cH:6][cH:5][c:4]1[CH2:3][CH:2]=[CH2:1]']
good = ['[OH:8][C:6](=[O:7])[CH:5]=[CH2:4].[CH3:1][C:2](=[CH2:3])[CH:18]=[CH2:17]>>[CH3:1][C:2]1=[CH:18][CH2:17][CH:5]([CH2:4][CH2:3]1)[C:6]([OH:8])=[O:7]',
        '[CH2:1]=[CH:2][CH2:3][O:17][c:16]1[cH:4][cH:5][cH:6][cH:14][cH:15]1>>[OH:17][c:16]1[cH:15][cH:14][cH:6][cH:5][c:4]1[CH2:1][CH:2]=[CH2:3]']

test = ['[CH3:18][N:19]([CH3:20])[c:21]1[cH:22][cH:23][cH:24][cH:25][cH:26]1.[CH2:1]=[CH:2][CH2:3][O:17][c:16]1[cH:4][cH:5][c:6]([cH:14][cH:15]1)[C:7]1=[N:8][NH:9][C:10](=[O:11])[CH2:12][CH2:13]1>>[OH:17][c:16]1[cH:15][cH:14][c:6]([cH:5][c:4]1[CH2:3][CH:2]=[CH2:1])[C:7]1=[N:8][NH:9][C:10](=[O:11])[CH2:12][CH2:13]1',
        '[CH3:15][C:14]1([CH3:16])[CH2:13][O:12][C:10](=[O:11])[CH:9]1[O:8][C:6](=[O:7])[CH:5]=[CH2:4].[CH3:1][C:2](=[CH2:3])[CH:18]=[CH2:17]>>[CH3:1][C:2]1=[CH:3][CH2:4][CH:5]([CH2:17][CH2:18]1)[C:6](=[O:7])[O:8][CH:9]1[C:10](=[O:11])[O:12][CH2:13][C:14]1([CH3:15])[CH3:16]']
```

```python
# loading and standardizing data
bad = [smiles(x) for x in bad]
good = [smiles(x) for x in good]
test = [smiles(x) for x in test]

for x in chain(bad, good, test):
    x.canonicalize()
    x.clean2d()
```

### Rules generation

Generated rules stored in ReactionContainer class. Now any reaction can use it.

```python
ReactionContainer.load_remapping_rules(zip(bad, good))  # pairs of bad and good mapping required.
```

### Fix procedure

Only appropriate rule will be applied 

```python
for x in test:
    print(x.fix_mapping() and 'fixed' or 'not fixed')  # Reactions have method for self-fixing
```
